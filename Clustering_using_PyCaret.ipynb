```python
## kmeans clutestering without any preprocessing
from pycaret.datasets import get_data
from pycaret.clustering import *

wineDataSet = get_data('wine', verbose = False)
setup(data = wineDataSet, verbose = False)

print("for cluster = 3")
x = create_model('kmeans', num_clusters = 3)
print("for cluster = 4")
x = create_model('kmeans', num_clusters = 4)
print("for cluster = 5")
x = create_model('kmeans', num_clusters = 5)
print("for cluster = 6")
x = create_model('kmeans', num_clusters = 6)
```

```python
## kmeans clustering with normalisation

setup(data = wineDataSet, normalize = True, normalize_method = 'zscore', verbose = False)
x = create_model('kmeans')
plot_model(x, plot = 'elbow')
```

```python
setup(data = wineDataSet, normalize = True, normalize_method = 'zscore', verbose = False)

print("for cluster c = 3")
x = create_model('kmeans', num_clusters = 3)
print("for cluster c = 4")
x = create_model('kmeans', num_clusters = 4)
print("for cluster c = 5")
x = create_model('kmeans', num_clusters = 5)
print("for cluster c = 6")
x = create_model('kmeans', num_clusters = 6)
```

```python
## kmeans using transformation
setup(data = wineDataSet, transformation = True, transformation_method = 'yeo-johnson', verbose=False)

print("For Cluster = 3")
x = create_model('kmeans', num_clusters = 3)

print("For Cluster = 4")
x = create_model('kmeans', num_clusters = 4)

print("For Cluster = 5")
x = create_model('kmeans', num_clusters = 5)
```

```python
setup(data = wineDataset, pca = True, pca_method = 'linear', verbose = False)
x = create_model('kmeans')
plot_model(x, plot = 'elbow')
```

```python
## kmeans with PCA
setup(data = wineDataSet, pca = True, pca_method = 'linear', verbose=False)

print("For Cluster = 3")
x = create_model('kmeans', num_clusters = 3)

print("For Cluster = 4")
x = create_model('kmeans', num_clusters = 4)

print("For Cluster = 5")
x = create_model('kmeans', num_clusters = 5)
```

```python
setup(data = wineDataSet, transformation = True,  normalize = True, normalize_method = 'zscore', transformation_method = 'yeo-johnson', verbose=False)
x = create_model('kmeans')
plot_model(x, plot = 'elbow')
```
